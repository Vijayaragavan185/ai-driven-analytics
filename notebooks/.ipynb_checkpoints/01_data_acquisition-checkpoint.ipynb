```python
# 01_data_acquisition.ipynb
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Sample data creation (for demonstration)
# In a real project, you might load data from CSV, API, or database
np.random.seed(42)
data = {
    'date': pd.date_range(start='2024-01-01', periods=100, freq='D'),
    'sales': np.random.normal(loc=1000, scale=200, size=100),
    'customer_count': np.random.poisson(lam=50, size=100),
    'marketing_spend': np.random.gamma(shape=5, scale=200, size=100)
}
```

```python
df = pd.DataFrame(data)

# Basic exploration
print("Dataset shape:", df.shape)
print("\nFirst 5 rows:")
print(df.head())
print("\nBasic statistics:")
print(df.describe())
```

```python
# Simple visualization
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(df['marketing_spend'], df['sales'])
plt.title('Marketing Spend vs. Sales')
plt.xlabel('Marketing Spend ($)')
plt.ylabel('Sales ($)')

plt.subplot(1, 2, 2)
plt.plot(df['date'], df['sales'])
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Save the data for future use
df.to_csv('../data/raw/sample_business_data.csv', index=False)
print("Data saved to '../data/raw/sample_business_data.csv'")
```
